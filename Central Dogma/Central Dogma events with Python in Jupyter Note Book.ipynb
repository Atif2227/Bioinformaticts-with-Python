```python
#Function for getting colored Nucleotides. 
#This won't here in Jupyter. Try VS code or any other IDE instead
def colored(seq):
    bcolors = {
        'A': '\033[92m',
        'C': '\033[94m',
        'G': '\033[93m',
        'T': '\033[91m',
        'U': '\033[91m',
        'reset': '\033[0;0m'
    }

    tmpStr = ""

    for nuc in seq:
        if nuc in bcolors:
            tmpStr += bcolors[nuc] + nuc
        else:
            tmpStr += bcolors['reset'] + nuc

    return tmpStr + '\033[0;0m'

print("Random DNA string - ", {colored(rand_DNA_str)})
print("Nucleotide Frequencies - ",{colored(count_nuc_freq(rand_DNA_str))})
```

## 5. Make a function for 'Reverse Transcription'. (Reverse Transcription is done by the enzyme reverse transcriptase and it generates cDNA. It is a process where mRNA is reverse transcribed into cDNA)

### Answer 5

```python
DNA_Reversecomplement = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C'}
def rev_trancription(self):
    '''Transcribes mRNA into cDNA'''
    return ''.join ([DNA_Reversecomplement[i] for i in self])[::-1]

print("The Reverse Complimentary DNA -", rev_trancription(rand_DNA_str))
```

## 7. Find the GC content of the DNA string.

### Answer 7

```python
def GC_content(self):
    '''Gives the GC content of the generated DNA string.'''
    return round((self.count('G') + self.count('C')) / len(self) *100)

print("GC content Percentage -", GC_content(rand_DNA_str),"%")
```

## 8. Find the GC content of subsection of fixed length from the DNA generated string.

### Answer 8

```python
def GC_content_subsec(self, k): #'k is the length of sebsection.'
    '''Returns the GC content from fixed lenth sub-section of the given DNA string.'''
    res = []
    for i in range(0, len(self) - k + 1, k):
        subseq = self[i:i + k]
        res.append(GC_content(subseq))
    return res

print("The GC content from the subsections - ", GC_content_subsec(rand_DNA_str,5))
```

```python
#Defining dictionaries for codon from DNA and RNA.

# Defining Codons
DNA_Codons = {
    # 'M' - START, '_' - STOP
    "GCT": "A", "GCC": "A", "GCA": "A", "GCG": "A",
    "TGT": "C", "TGC": "C",
    "GAT": "D", "GAC": "D",
    "GAA": "E", "GAG": "E",
    "TTT": "F", "TTC": "F",
    "GGT": "G", "GGC": "G", "GGA": "G", "GGG": "G",
    "CAT": "H", "CAC": "H",
    "ATA": "I", "ATT": "I", "ATC": "I",
    "AAA": "K", "AAG": "K",
    "TTA": "L", "TTG": "L", "CTT": "L", "CTC": "L", "CTA": "L", "CTG": "L",
    "ATG": "M",
    "AAT": "N", "AAC": "N",
    "CCT": "P", "CCC": "P", "CCA": "P", "CCG": "P",
    "CAA": "Q", "CAG": "Q",
    "CGT": "R", "CGC": "R", "CGA": "R", "CGG": "R", "AGA": "R", "AGG": "R",
    "TCT": "S", "TCC": "S", "TCA": "S", "TCG": "S", "AGT": "S", "AGC": "S",
    "ACT": "T", "ACC": "T", "ACA": "T", "ACG": "T",
    "GTT": "V", "GTC": "V", "GTA": "V", "GTG": "V",
    "TGG": "W",
    "TAT": "Y", "TAC": "Y",
    "TAA": "_", "TAG": "_", "TGA": "_"
}

RNA_Codons = {
    # 'M' - START, '_' - STOP
    "GCU": "A", "GCC": "A", "GCA": "A", "GCG": "A",
    "UGU": "C", "UGC": "C",
    "GAU": "D", "GAC": "D",
    "GAA": "E", "GAG": "E",
    "UUU": "F", "UUC": "F",
    "GGU": "G", "GGC": "G", "GGA": "G", "GGG": "G",
    "CAU": "H", "CAC": "H",
    "AUA": "I", "AUU": "I", "AUC": "I",
    "AAA": "K", "AAG": "K",
    "UUA": "L", "UUG": "L", "CUU": "L", "CUC": "L", "CUA": "L", "CUG": "L",
    "AUG": "M",
    "AAU": "N", "AAC": "N",
    "CCU": "P", "CCC": "P", "CCA": "P", "CCG": "P",
    "CAA": "Q", "CAG": "Q",
    "CGU": "R", "CGC": "R", "CGA": "R", "CGG": "R", "AGA": "R", "AGG": "R",
    "UCU": "S", "UCC": "S", "UCA": "S", "UCG": "S", "AGU": "S", "AGC": "S",
    "ACU": "T", "ACC": "T", "ACA": "T", "ACG": "T",
    "GUU": "V", "GUC": "V", "GUA": "V", "GUG": "V",
    "UGG": "W",
    "UAU": "Y", "UAC": "Y",
    "UAA": "_", "UAG": "_", "UGA": "_"
}

#Now Translate the DNA/RNA sequence.
def translate_seq(self, init_pos=0):
    """Translates a DNA sequence into an aminoacid sequence"""
    return [DNA_Codons[self[i:i + 3]] for i in range(init_pos, len(self) - 2, 3)]

print("The Amino Acid sequence from the DNA string is -", translate_seq(rand_DNA_str))
```

## 10. Calculate the proportion/frequency of asked amino acid present in the AA sequence.

### Answer 10

```python
from typing import Counter

def codon_usage(self, aminoacid):
    """Provides the frequency of asked aminoacid with its codon in a DNA sequence"""
    tmpList = []
    for i in range(0, len(self) - 2, 3):
        if DNA_Codons[self[i:i + 3]] == aminoacid:
            tmpList.append(self[i:i + 3])

    freqDict = dict(Counter(tmpList))
    totalWight = sum(freqDict.values())
    for self in freqDict:
        freqDict[self] = round(freqDict[self] / totalWight, 2)
    return freqDict

print("The frequency of Amino acid A is -", codon_usage(rand_DNA_str,"A"))
```

## 11. Generate the 6 reading frames from the DNA Sequence.

### Answer 11

```python
def gen_RF(self):
    """Generate the six reading frames of a DNA sequence."""
    frames = []
    frames.append(translate_seq(self,0))
    frames.append(translate_seq(self,1))
    frames.append(translate_seq(self,2))
    frames.append(translate_seq(rev_trancription(self),0))
    frames.append(translate_seq(rev_trancription(self),1))
    frames.append(translate_seq(rev_trancription(self),2))
    return frames

print("The reading frames - ")
for i in gen_RF(rand_DNA_str):
    print(i)
```

## 12. Extract all possible proteins in an aminoacid seq and return a list of possible proteins

### Answer 12

```python
def proteins_from_rf(aa_seq):
    """Compute all possible proteins in an aminoacid seq and return a list of possible proteins"""
    current_prot = []
    proteins = []
    for aa in aa_seq:
        if aa == "_":
            # STOP accumulating amino acids if _ - STOP was found
            if current_prot:
                for p in current_prot:
                    proteins.append(p)
                current_prot = []
        else:
            # START accumulating amino acids if M - START was found
            if aa == "M":
                current_prot.append("")
            for i in range(len(current_prot)):
                current_prot[i] += aa
    return proteins
```
